**Univariate Analysis**

**Let's check the distribution for numerical columns**

**Observations on Distance**

```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.float_format", lambda x: "%.3f" %x) #to restrict any float number to 3dp
```

```python
#accessing my file from my google drive
from google.colab import drive
drive.mount("/content/drive")
```

```python
#importing the dataset to be utilized
path= "/content/drive/MyDrive/Week 1  Foundations Python and Statistics/Melbourne_Housing.csv"
data=pd.read_csv(path)
```

```python
sns.histplot(data=data, x="Distance", stat="density")
plt.show()
sns.boxplot(data=data, x="Distance")
plt.show()
```

.The distribution is skewed towards right

.There are many outliers present in this column

.Values above 25kms are being represented as outliers in the boxplot, indicating there are many properties that are at least 25kms away from CBD

**Observations on Landsize**

```python
sns.displot(data=data, x="Landsize", kind="kde")
plt.show()
sns.boxplot(data=data, x="Landsize")
plt.show()
```

**Types of univariate analysis include 1) graphical, 2) table and 3) descriptive statistics.**

```python
#converting Landsize to sq. Kilometers from sq. metres
sns.displot(data=data, x=data["Landsize"]/1000000, kind="kde")
plt.show()
sns.boxplot(data=data, x=data["Landsize"]/1000000)
plt.show()
```

```python
data.loc[data["Landsize"]>60000]
```

**Observations on BuildingArea**

```python
data['BuildingArea']=pd.to_numeric(data['BuildingArea'], errors='coerce')
```

```python
sns.displot(data=data, x="BuildingArea", kind="kde")
plt.show()
sns.boxplot(data=data, x="BuildingArea")
plt.show()
```

**Observations on Price**
